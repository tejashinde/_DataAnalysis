```python
%matplotlib inline
import pandas
import numpy
import matplotlib.pyplot as pyplot
```

```python
dataframe = pandas.read_csv('./tmp/all_stocks_5yr.csv' , parse_dates = ['date'])
```

```python
dataframe.set_index('date')
```

```python
dataframe_AAPL = dataframe[dataframe.Name == 'AAPL']
dataframe_ZTS = dataframe[dataframe.Name == 'ZTS']
```

```python
df3 = pandas.merge(dataframe_AAPL , dataframe_ZTS , how = 'outer' , on = 'date').set_index('date')
```

```python
df3.resample('4M').sum().plot()
```

```python
dataframe[['high','low']].apply(lambda x : numpy.mean((x[0],x[1])) , axis = 1)
```

```python
dataframe['Volume Change'] = dataframe.volume.pct_change()
```

```python
company_df = dataframe.pivot(index = 'date' , columns = 'Name' , values = 'volume').fillna(method = 'bfill').pct_change()
```

```python
company_df[['ALLE','AAP']].plot()
```
