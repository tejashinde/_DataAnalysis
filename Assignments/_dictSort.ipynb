```python
import pandas
```

```python
dictionary = {"Name" : ["Tejas","Aditya","Prathamesh","Sarthak","Dharmendra"] , "PRN" : [37 , 21 , 11 , 29 , 1] , "Age" : [21 , 26 , 23 , 18 , 90]}
```

```python
pandas.DataFrame(sorted(zip(dictionary["PRN"] , dictionary["Name"] , dictionary["Age"]))).rename(columns = {0 : "PRN" , 1 : "Name" , 2 : "Age"})
```
